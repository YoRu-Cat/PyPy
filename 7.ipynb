## Question 1: Population, Fitness, Selection

```python
population = ["0011", "0101", "1001", "1110"]
rows = []
for ch in population:
    x = int(ch, 2)
    fit = x ** 2
    rows.append((ch, x, fit))
ranked = sorted(rows, key=lambda t: t[2], reverse=True)
parents = ranked[:2]
print("Chromosome | x | fitness")
for r in rows:
    print(r[0], r[1], r[2])
print("\nRanked (best to worst):")
for r in ranked:
    print(r)
print("\nSelected parents:")
for p in parents:
    print(p)
```

## Question 2: Crossover / New Generation

```python
dist = {
    ('A','B'):10, ('A','C'):15, ('A','D'):20,
    ('B','C'):35, ('B','D'):25, ('C','D'):30
}

def d(a, b):
    if (a, b) in dist:
        return dist[(a, b)]
    return dist[(b, a)]

def route_cost(route):
    return sum(d(route[i], route[i+1]) for i in range(len(route)-1))

routes = [
    ['A','B','C','D'],
    ['A','C','B','D'],
    ['B','A','D','C'],
    ['C','B','A','D']
]

scored = [(r, route_cost(r)) for r in routes]
ranked = sorted(scored, key=lambda x: x[1])
p1, p2 = ranked[0][0], ranked[1][0]

point = 2
def make_child(a, b, point):
    head = a[:point]
    tail = [x for x in b if x not in head]
    return head + tail

child1 = make_child(p1, p2, point)
child2 = make_child(p2, p1, point)
new_gen = [p1, p2, child1, child2]

print("Route | Distance")
for r, c in scored:
    print(' -> '.join(r), c)
print("\nBest two parents:")
print('P1:', ' -> '.join(p1), route_cost(p1))
print('P2:', ' -> '.join(p2), route_cost(p2))
print("\nOffspring:")
print('C1:', ' -> '.join(child1), route_cost(child1))
print('C2:', ' -> '.join(child2), route_cost(child2))
print("\nNew generation:")
for r in new_gen:
    print(' -> '.join(r))
```

## Question 3: Mutation

```python
import random

random.seed(7)
population = [
    ['A','B','C','D'],
    ['B','C','A','D'],
    ['D','A','B','C'],
    ['C','B','D','A']
]

def mutate_swap(ch):
    ch = ch[:]
    i, j = random.sample(range(len(ch)), 2)
    ch[i], ch[j] = ch[j], ch[i]
    return ch

mutated = [mutate_swap(ch) for ch in population]

print("Original chromosomes:")
for ch in population:
    print(' '.join(ch))
print("\nMutated chromosomes:")
for ch in mutated:
    print(' '.join(ch))
```

## Question 4: GA Problem Solving

```python
import random

random.seed(2)

def expr(x, y, z):
    return 6*(x**3) + 9*(y**2) + 90*z

def fitness(sol):
    x, y, z = sol
    return 1 / (abs(expr(x, y, z) - 25) + 1e-9)

population = [
    (1.0, 1.0, 0.0),
    (2.0, 1.0, 0.0),
    (1.0, 2.0, 0.0),
    (2.0, 2.0, 0.0)
]

print("Initial evaluation:")
for sol in population:
    val = expr(*sol)
    fit = fitness(sol)
    print(sol, "expr=", round(val, 4), "fitness=", round(fit, 6))

gens = 20
for _ in range(gens):
    ranked = sorted(population, key=fitness, reverse=True)
    p1, p2 = ranked[0], ranked[1]
    base = [p1, p2]
    new_pop = []
    for _ in range(4):
        src = random.choice(base)
        x, y, z = src
        x += random.choice([-0.1, 0.1])
        y += random.choice([-0.1, 0.1])
        z += random.choice([-0.1, 0.1])
        new_pop.append((round(x, 3), round(y, 3), round(z, 3)))
    population = new_pop

best = max(population, key=fitness)
print("\nAfter generations:")
print("Best solution:", best)
print("Expression value:", round(expr(*best), 6))
print("Distance from 25:", round(abs(expr(*best)-25), 6))
```
