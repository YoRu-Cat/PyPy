## Question 1

```python
import pandas as pd

try:
    df = pd.read_csv('customers.csv')
except Exception:
    df = pd.DataFrame({
        'CustomerID': ['C1','C2','C3','C2'],
        'Age': ['24', None, '35', None],
        'Gender': ['Male', 'Female', None, 'Female'],
        'Annual_Income': ['50,000', 'not available', '72,500', 'not available'],
        'City': ['Lahore', None, 'Karachi', None]
    })

df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Annual_Income'] = pd.to_numeric(df['Annual_Income'].astype(str).str.replace(',', '', regex=False), errors='coerce')
df = df[df.isna().sum(axis=1) <= 2].copy()
df['Age'] = df['Age'].fillna(df['Age'].median())
df['City'] = df['City'].fillna(df['City'].mode().iloc[0])
df = df.drop_duplicates()
df['Gender'] = df['Gender'].fillna(df['Gender'].mode().iloc[0])
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})

print(df.isna().sum())
print(df)
```

## Question 2

```python
import pandas as pd

try:
    df = pd.read_csv('patients.csv')
except Exception:
    df = pd.DataFrame({
        'Patient_ID': [1,2,3,4],
        'Admission_Date': ['01-02-2024','2024/03/01','invalid',None],
        'Discharge_Date': ['05-02-2024','2024/03/10','2024/03/15','invalid'],
        'Severity': ['Low','Medium','High','Low'],
        'Department': ['Cardiology','Neurology','Orthopedic','Cardiology']
    })

df['Admission_Date'] = pd.to_datetime(df['Admission_Date'], errors='coerce', dayfirst=True)
df['Discharge_Date'] = pd.to_datetime(df['Discharge_Date'], errors='coerce', dayfirst=True)
df['Severity'] = df['Severity'].map({'Low': 1, 'Medium': 2, 'High': 3})
df = pd.get_dummies(df, columns=['Department'], prefix='Department')
df['Severity'] = pd.to_numeric(df['Severity'], errors='coerce')

print(df.dtypes)
print(df)
```

## Question 3

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler

try:
    df = pd.read_csv('housing.csv')
except Exception:
    df = pd.DataFrame({
        'Area': [500, 900, 1200, 2000, 3500, 5000],
        'Bedrooms': [1, 2, 2, 3, 4, 5],
        'Price': [50, 85, 120, 180, 300, 420],
        'Distance_from_City': [40, 30, 22, 12, 8, 5],
        'Age_of_House': [30, 20, 15, 10, 6, 2]
    })

X = df[['Area', 'Bedrooms', 'Distance_from_City', 'Age_of_House']]
y = df['Price']

to_normalize = ['Area', 'Distance_from_City']
to_standardize = ['Age_of_House']
print('Normalize:', to_normalize)
print('Standardize:', to_standardize)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

X_train = X_train.reset_index(drop=True)
X_test = X_test.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)

mm = MinMaxScaler()
ss = StandardScaler()

X_train[to_normalize] = mm.fit_transform(X_train[to_normalize])
X_test[to_normalize] = mm.transform(X_test[to_normalize])

X_train[to_standardize] = ss.fit_transform(X_train[to_standardize])
X_test[to_standardize] = ss.transform(X_test[to_standardize])

print('Train')
print(X_train)
print('\nTest')
print(X_test)
```

## Question 4

```python
import pandas as pd

try:
    df = pd.read_csv('orders.csv')
except Exception:
    df = pd.DataFrame({
        'Order_ID': [1, 2, 3, 2],
        'Restaurant_Name': ['R1', 'R2', 'R3', 'R2'],
        'Cuisine_Type': ['Italian', 'Chinese', 'Pakistani', 'Chinese'],
        'Delivery_Time': [30, None, 45, None],
        'Customer_Rating': [4.5, 4.0, None, 4.0]
    })

dup_count = df.duplicated().sum()
df = df.drop_duplicates()
df['Delivery_Time'] = pd.to_numeric(df['Delivery_Time'], errors='coerce')
df['Customer_Rating'] = pd.to_numeric(df['Customer_Rating'], errors='coerce')
df['Delivery_Time'] = df['Delivery_Time'].fillna(df['Delivery_Time'].mean())
df['Customer_Rating'] = df['Customer_Rating'].fillna(df['Customer_Rating'].median())
df = pd.get_dummies(df, columns=['Cuisine_Type'], prefix='Cuisine')

print(dup_count)
print(df)
```

## Question 5

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

try:
    df = pd.read_csv('students.csv')
except Exception:
    df = pd.DataFrame({
        'Name': ['Ali', 'Sara', 'John', 'Ali'],
        'Marks': ['85 marks', None, '72 marks', '85 marks'],
        'Grade': ['A', 'B', 'C', 'A'],
        'Study_Hours': [5, 4, 3, 5],
        'Passed': ['Yes', 'Yes', 'No', 'Yes']
    })

df['Marks'] = pd.to_numeric(df['Marks'].astype(str).str.extract('(\\d+)', expand=False), errors='coerce')
df['Marks'] = df['Marks'].fillna(df['Marks'].mean())
df = df.drop_duplicates()
df['Grade'] = df['Grade'].map({'A': 4, 'B': 3, 'C': 2, 'D': 1})
le = LabelEncoder()
df['Passed'] = le.fit_transform(df['Passed'])
ss = StandardScaler()
df[['Study_Hours']] = ss.fit_transform(df[['Study_Hours']])

print(df)
```

## Question 6

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

data = {
    'Student_ID': [1, 2, 3, 4, 2],
    'Course': ['AI', 'Web Dev', 'AI', None, 'Web Dev'],
    'Completion_Percentage': ['80', '90%', None, '70', '90%'],
    'Enroll_Date': ['2024-01-10', '10/02/2024', 'invalid', '2024-03-01', '10/02/2024'],
    'Feedback': ['Good', 'Excellent', 'Average', 'Good', 'Excellent']
}
df = pd.DataFrame(data)

print('Missing before cleaning:')
print(df.isna().sum())

df = df.drop_duplicates()
df['Completion_Percentage'] = pd.to_numeric(df['Completion_Percentage'].astype(str).str.replace('%', '', regex=False), errors='coerce')
df['Enroll_Date'] = pd.to_datetime(df['Enroll_Date'], errors='coerce', dayfirst=True)
df['Completion_Percentage'] = df['Completion_Percentage'].fillna(df['Completion_Percentage'].mean())
df['Course'] = df['Course'].fillna(df['Course'].mode().iloc[0])
df = pd.get_dummies(df, columns=['Course'], prefix='Course')
le = LabelEncoder()
df['Feedback'] = le.fit_transform(df['Feedback'])
mm = MinMaxScaler()
df[['Completion_Percentage']] = mm.fit_transform(df[['Completion_Percentage']])

print('\nMissing after cleaning:')
print(df.isna().sum())
print('\nFinal data:')
print(df)
```

## Question 7

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

data = {
    'Product_ID': [101, 102, 103, 104, 101],
    'Category': ['Electronics', 'Clothing', 'Electronics', 'Grocery', 'Electronics'],
    'Price': ['1000', '2000', 'invalid', '500', '1000'],
    'Quantity_Sold': [5, 10, None, 8, 5],
    'Discount': [10, 20, 15, None, 10],
    'Purchase_Date': ['2024/01/01', '01-02-2024', '2024-03-01', 'invalid', '2024/01/01']
}
df = pd.DataFrame(data)

df = df.drop_duplicates()
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'], errors='coerce', dayfirst=True)
df['Quantity_Sold'] = pd.to_numeric(df['Quantity_Sold'], errors='coerce').fillna(pd.to_numeric(df['Quantity_Sold'], errors='coerce').median())
df['Discount'] = pd.to_numeric(df['Discount'], errors='coerce').fillna(pd.to_numeric(df['Discount'], errors='coerce').mean())
df['Price'] = df['Price'].fillna(df['Price'].mean())
df = pd.get_dummies(df, columns=['Category'], prefix='Category')

X = df.drop(columns=['Product_ID', 'Purchase_Date'])
y = pd.Series([0] * len(X))
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
ss = StandardScaler()
X_train = X_train.copy()
X_test = X_test.copy()
X_train[['Price', 'Quantity_Sold']] = ss.fit_transform(X_train[['Price', 'Quantity_Sold']])
X_test[['Price', 'Quantity_Sold']] = ss.transform(X_test[['Price', 'Quantity_Sold']])

print(X_train)
print(X_test)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

df = pd.DataFrame({
    'Customer_ID': [1,2,3,4,2],
    'Age': ['25', 'unknown', None, '40', 'unknown'],
    'Income': ['50,000', '75,000', '60,000', None, '75,000'],
    'Loan_Status': ['Approved', 'Rejected', 'Approved', 'Rejected', 'Rejected'],
    'Credit_Score': [700, None, 650, 620, None],
    'City': ['Lahore', 'Karachi', None, 'Islamabad', 'Karachi']
})

print('Missing before cleaning:')
print(df.isna().sum())

df = df.drop_duplicates()
df['Income'] = pd.to_numeric(df['Income'].astype(str).str.replace(',', '', regex=False), errors='coerce')
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Credit_Score'] = pd.to_numeric(df['Credit_Score'], errors='coerce').fillna(pd.to_numeric(df['Credit_Score'], errors='coerce').mean())
df['City'] = df['City'].fillna(df['City'].mode().iloc[0])
df['Income'] = df['Income'].fillna(df['Income'].mean())
le = LabelEncoder()
df['Loan_Status'] = le.fit_transform(df['Loan_Status'])
df = pd.get_dummies(df, columns=['City'], prefix='City')
ss = StandardScaler()
df[['Income', 'Credit_Score']] = ss.fit_transform(df[['Income', 'Credit_Score']])

print('\nMissing after cleaning:')
print(df.isna().sum())
print('\nFinal data:')
print(df)
```
