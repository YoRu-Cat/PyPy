```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
print("Q1: Mean, Median, Mode, Range")

data = {'Day': ['Mon','Tue','Wed','Thu','Fri','Sat','Sun'],
        'Temperature': [30, 32, 28, 35, 30, 33, 29],
        'Rainfall': [5, 0, 12, 3, 8, 0, 6]}
df = pd.DataFrame(data)
print(df)

print("\nMean:", df['Temperature'].mean())
print("Median:", df['Temperature'].median())
print("Mode:", df['Temperature'].mode()[0])
print("Range:", df['Temperature'].max() - df['Temperature'].min())
```

```python
print("Q2: Types of Attributes")

data = {'PatientID': [101, 102, 103, 104, 105],
        'Age': [25, 40, 35, 60, 50],
        'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
        'BloodPressure': [120, 130, 110, 140, 125],
        'Diagnosis': ['Flu', 'Diabetes', 'Flu', 'Heart Disease', 'Diabetes']}
df2 = pd.DataFrame(data)
print(df2)

print("\nNominal: PatientID, Gender, Diagnosis")
print("Ordinal: None")
print("Interval/Ratio: Age, BloodPressure")
```

```python
print("Q3: Dataset Repositories")

iris = load_iris()
df3 = pd.DataFrame(iris.data, columns=iris.feature_names)
df3['species'] = iris.target

print(df3.head())
print("\nRows:", df3.shape[0])
print("Columns:", df3.shape[1])
print("Column Names:", list(df3.columns))
```

```python
print("Q4: Properties of Attributes")

data = {'CustomerID': [1, 2, 3, 4, 5],
        'AccountBalance': [15000, 23000, 5000, 42000, 18000],
        'LoanStatus': ['Approved', 'Pending', 'Rejected', 'Approved', None],
        'Age': [28, 45, 33, 50, 38]}
df4 = pd.DataFrame(data)
print(df4)

print("\nMin:", df4['AccountBalance'].min())
print("Max:", df4['AccountBalance'].max())
print("Sum:", df4['AccountBalance'].sum())
print("\nCumulative Sum:\n", df4['AccountBalance'].cumsum())
print("\nNon-Null Count:\n", df4.count())
```

```python
print("Q5: Iris Flower Dataset")

try:
	pd
except NameError:
	import pandas as pd

df5 = pd.DataFrame(iris.data, columns=iris.feature_names)
names = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
df5['species'] = [names[t] for t in iris.target]

print(df5.head(100))
print("\nFeatures:", iris.feature_names)
print("Target:", list(names.values()))
print("\nMean Petal Length per Species:")
print(df5.groupby('species')['petal length (cm)'].mean())
```

```python
print("Q6: Line Plot - Daily Steps Tracker")

days = [1, 2, 3, 4, 5, 6, 7]
steps = [5000, 7000, 8000, 6500, 9000, 10000, 7500]

plt.figure(figsize=(8, 5))
plt.plot(days, steps, marker='o')
plt.xlabel('Day')
plt.ylabel('Steps')
plt.title('Daily Steps Tracker')
plt.grid(True)
plt.show()
```

```python
print("Q7: Scatter Plot - Car Age vs Rental Price")

car_age = [1, 3, 5, 2, 8, 4, 6]
rental_price = [50, 45, 40, 48, 35, 42, 38]

plt.figure(figsize=(8, 5))
plt.scatter(car_age, rental_price, color='red', s=100)
plt.xlabel('Car Age (years)')
plt.ylabel('Rental Price ($)')
plt.title('Car Age vs Rental Price')
plt.grid(True)
plt.show()

print("Newer cars have higher rental prices (negative correlation).")
```

```python
print("Q8: Bar Chart - Fruit Shop")

fruits = ["Apples", "Bananas", "Cherries", "Dates"]
quantity = [50, 75, 20, 40]

plt.figure(figsize=(8, 5))
plt.bar(fruits, quantity, color=['red', 'yellow', 'darkred', 'brown'])
plt.xlabel('Fruit')
plt.ylabel('Quantity Sold')
plt.title('Fruit Shop - Daily Sales')
plt.grid(axis='y')
plt.show()
```

```python
print("Q9: Histogram - Student Test Scores")

scores = np.random.normal(70, 10, 100)

plt.figure(figsize=(8, 5))
plt.hist(scores, bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Score')
plt.ylabel('Number of Students')
plt.title('Distribution of Student Test Scores')
plt.show()

print("Most students scored between 60-80.")
```

```python
print("Q10a: Pie Chart - Sales by Region")

sales = [300, 250, 150, 100]
regions = ["North", "South", "East", "West"]

plt.figure(figsize=(7, 7))
plt.pie(sales, labels=regions, autopct='%1.1f%%', startangle=90,
        colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.legend(regions, title="Regions")
plt.title('Sales Distribution by Region')
plt.show()
```

```python
print("Q10b: Box Plot - Employee Salaries")

salaries = [2500, 3000, 4000, 3500, 4500, 5000, 3200]

plt.figure(figsize=(6, 6))
plt.boxplot(salaries, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.ylabel('Salary ($)')
plt.title('Employee Salary Distribution')
plt.show()

print("Min:", min(salaries), "| Max:", max(salaries))
print("Median:", np.median(salaries))
print("Q1:", np.percentile(salaries, 25), "| Q3:", np.percentile(salaries, 75))
```
