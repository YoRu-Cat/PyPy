```python
#Ex 1.2
fruits = ['apple', 'banana', 'cherry', 'date', 'fig']
fruits[1] = 'blueberry'
fruits.pop()
fruits.append('grape')

print(fruits)
```

```python
#Ex 1.3
numbers = [10, 5, 7, 3, 9]

print("Sum:", sum(numbers))
print("Maximum:", max(numbers))
print("Minimum:", min(numbers))

numbers.sort()
print("Sorted list:", numbers)
```

```python
#Ex 1.4
squares = [n**2 for n in range(1, 11) if n % 2 == 0]
print(squares)

fruits = ['apple', 'banana', 'cherry', 'date', 'fig']
short_fruits = [fruit for fruit in fruits if len(fruit) < 6]
print(short_fruits)
```

```python
#Ex 1.5
colors = ('red', 'green', 'blue', 'yellow')
print(colors[0])
print(colors[-1])
print(colors[1:3])

colors[1] = 'black'
```

```python
#Ex 1.7
student = {'name': 'Ali', 'age': 20, 'grade': 'A'}
print(student['name'])
student['city'] = 'Lahore'
student['grade'] = 'A+'
del student['age']

print(student)
```

```python
#Ex 1.8
scores = {'Math': 90, 'Physics': 85, 'Chemistry': 92}
print(scores.keys())
print(scores.values())
print("Total:", sum(scores.values()))
print(f"Ali scored {scores['Math']} in Math")
```

```python
#Ex 1.9
tools = {'Photoshop', 'Illustrator', 'Acrobat'}
x = ['InDesign','Bridge']
tools.add('Premiere')
tools.remove('Acrobat')
print(tools)
tools.clear()

print(tools)
tools.update(x)
print(tools)
```

```python
#Ex 1.10
DS_skills = {'Python', 'R', 'SQL', 'Git'}
DE_skills = {'Python', 'Java', 'SQL', 'Hadoop'}
print("Union:", DS_skills | DE_skills)
print("Intersection:", DS_skills & DE_skills)
print("Difference:", DS_skills - DE_skills)
print("Symmetric Difference:", DS_skills ^ DE_skills)
```
