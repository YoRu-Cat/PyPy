```python
#Q3
age = int(input("Enter your age: "))
is_student = input("Are you a student? (True/False): ").strip().lower() == "true"

if age >= 18 and age <= 60 and is_student:
    print("\nBetween 18-60 AND a student: True")
else:
    print("\nBetween 18-60 AND a student: False")

if not is_student or age < 18:
    print("NOT a student OR younger than 18: True")
else:
    print("NOT a student OR younger than 18: False")
```

```python
#Q4
def getGrade(ans):
    if ans >= 90 and ans <= 100:
        return 'A'
    elif ans >= 80 and ans <= 89:
        return 'B'
    elif ans >= 70 and ans <= 79:
        return 'C'
    elif ans >= 60 and ans <= 69:
        return 'D'
    else:
        return 'F'

x = int(input('Enter you marks: '))
grade = getGrade(x)
print(grade)
```

```python
#Q5
while True:
    number = int(input("Enter a number (0 to stop): "))
    if number < 0:
        print("Negative numbers are not allowed.")
        break
    if number == 0:
        print("Exiting program.")
        break
    print("You entered: ",number)
```

```python
#Q6
print("Multiplication Table of 7:")
for i in range(1, 11):
    result = 7 * i
    print(f"7 × {i} = {result}")
```

```python
#Q7
print("Odd numbers from 1 to 15:")
for num in range(1, 16, 2):
    print(num, end=" ")
print('\n')
print("Even numbers from 20 down to 2:")
for num in range(20, 0, -2):
    print(num, end=" ")
```
