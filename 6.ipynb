## Q1: Alpha–Beta Pruning in a Game Decision System

```python
import math

def alphabeta(tree, depth, alpha, beta, maximizing, path="Root"):
    if isinstance(tree, int):
        print(f"  Leaf {path} = {tree}")
        return tree

    if maximizing:
        value = -math.inf
        for i, child in enumerate(tree):
            child_path = f"{path}->{'L' if i==0 else 'R'}"
            print(f"MAX node {path}: \u03b1={alpha}, \u03b2={beta}")
            score = alphabeta(child, depth-1, alpha, beta, False, child_path)
            value = max(value, score)
            alpha = max(alpha, value)
            if alpha >= beta:
                print(f"  ** Pruned at {path} (\u03b1={alpha} >= \u03b2={beta})")
                break
        return value
    else:
        value = math.inf
        for i, child in enumerate(tree):
            child_path = f"{path}->{'L' if i==0 else 'R'}"
            print(f"MIN node {path}: \u03b1={alpha}, \u03b2={beta}")
            score = alphabeta(child, depth-1, alpha, beta, True, child_path)
            value = min(value, score)
            beta = min(beta, value)
            if alpha >= beta:
                print(f"  ** Pruned at {path} (\u03b1={alpha} >= \u03b2={beta})")
                break
        return value

# Terminal values: 3,5,6,9,1,2,0,-1 in a 3-level tree
game_tree = [[[3, 5], [6, 9]], [[1, 2], [0, -1]]]

result = alphabeta(game_tree, 3, -math.inf, math.inf, True)
print(f"\nOptimal value at root: {result}")
```

## Q2: Hill Climbing for Route Elevation Optimization

```python
def elevation(x):
    return -(x - 3)**2 + 9

x = 0
path = [x]

while True:
    neighbors = []
    if x - 1 >= 0: neighbors.append(x - 1)
    if x + 1 <= 6: neighbors.append(x + 1)

    best = max(neighbors, key=elevation)
    if elevation(best) <= elevation(x):
        break
    x = best
    path.append(x)

print("Positions visited:", path)
print(f"Final position: x = {x}, Max elevation: {elevation(x)}")
```

## Q3: Hill Climbing with Local Maximum in Resource Allocation

```python
productivity = {'A': 3, 'B': 6, 'C': 5, 'D': 7, 'E': 9, 'F': 8}
neighbors = {'A': ['B', 'C'], 'B': ['D'], 'C': ['D'], 'D': ['E', 'F'], 'E': [], 'F': []}

current = 'A'
path = [current]

while True:
    nbrs = neighbors[current]
    if not nbrs:
        break
    best = max(nbrs, key=lambda s: productivity[s])
    if productivity[best] <= productivity[current]:
        break
    current = best
    path.append(current)

print("States visited:", " -> ".join(path))
print(f"Final state: {current}, Productivity: {productivity[current]}")


print("\nHill climbing can get stuck at a local maximum when no neighbor" 
"      \nhas a higher value, even if a better state exists elsewhere. " 
"      \nHere the algorithm reaches E (the global max) via A->B->D->E," 
"      \nbut with different graph structures it could stop at a local peak.")
```
